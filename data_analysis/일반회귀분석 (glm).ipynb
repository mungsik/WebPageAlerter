```python
import pandas as pd
df = pd.read_csv("datas/union_datas.csv")
df.shape
```

```python
df.head()
```

```python
## fluctuation 0을 기준으로 1, 0으로 new_fluctuation 구분하기

for i in range(len(df)):
    if df.loc[i, 'fluctuation'] >= 0:
        df.loc[i, 'new_fluctuation'] = "1"
    else:
        df.loc[i, 'new_fluctuation'] = "0"

df
```

```python
# statsmodels package loading
import statsmodels.api as sm

# logistic regression
feature = df["check_contract"]
output = df["new_fluctuation"]
model = sm.formula.glm("output ~ feature", df, family = sm.families.Binomial()).fit()
print(model.summary())
```
